#dropna()

```python
import pandas as pd

health_data = pd.read_csv("health_data.csv", header=0, sep=",")
#dropna()function to remove the specific. axis=0 means that we want to remove all rows that have a specific value:
health_data.dropna(axis=0,inplace=True)

print(health_data)
```
